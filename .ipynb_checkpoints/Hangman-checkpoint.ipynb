```python
import random

words = ["boots", "horse", "trail", "cowboy", "west", "cactus"]

word = random.choice(words)

incorrect = []

guessed = []

turns=7

for i in range(len(word)):
    guessed.append("_")
print(*[i for i in guessed])

while(turns):
    
    if turns == 7:      
        print("H A N G M A N")
        print("")
        print("\n"*2)
    elif turns == 6:
        print("H A N G M A N")
        print("")
        print("   +---+")
        print("       |")
        print("       |")
        print("       |")
        print("       |")
        print("       |")
        print("  ======")
        print("\n"*2)        
    elif turns == 5:
        print("H A N G M A N")
        print("")        
        print("   +---+")
        print("   |   |")
        print("   O   |")
        print("       |")
        print("       |")
        print("       |")
        print("  ======")
        print("\n"*2)        
    elif turns == 4:
        print("H A N G M A N")   
        print("")        
        print("   +---+")
        print("   |   |")
        print("   O   |")
        print("   |   |")
        print("       |")
        print("       |")
        print("  ======")
        print("\n"*2)        
    elif turns == 3:        
        print("H A N G M A N")
        print("")        
        print("   +---+")
        print("   |   |")
        print("   O   |")
        print("  /|   |")
        print("       |")
        print("       |")
        print("  ======")
        print("\n"*2)        
    elif turns == 2:   
        print("H A N G M A N")   
        print("")        
        print("   +---+")
        print("   |   |")
        print("   O   |")
        print("  /|\  |")
        print("       |")
        print("       |")
        print("  ======")
        print("\n"*2)        
    elif turns == 1:
        print("H A N G M A N") 
        print("")        
        print("   +---+")
        print("   |   |")
        print("   O   |")
        print("  /|\  |")
        print("  /    |")
        print("       |")
        print("  ======")
        print("\n"*2)        



    letter_guess = input("Guess a letter:")
    if(letter_guess in incorrect):
        print("You already tried that")
        turns += 1

    if letter_guess in word:
        for i in range(len(word)):
            if word[i] == letter_guess:
                guessed[i] = word[i] 
    else:
        print("Incorrect")
        incorrect.append(letter_guess)
        turns -= 1
    
    guess_word = [i for i in guessed]
    guess_word = "".join(guess_word)

    if word == guess_word:
        print("YOU WIN!")
        break
    print(*([i for i in guessed]))
    if turns == 0:  
        print("H A N G M A N")
        print("")        
        print("   +---+")
        print("   |   |")
        print("   O   |")
        print("  /|\  |")
        print("  / \  |")
        print("       |")
        print("  ======")
        print("")
        print("Sorry, the correct word was", word,"\n  -- GAME OVER --")
        exit()
```
